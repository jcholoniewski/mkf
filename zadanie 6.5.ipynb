```python
import numpy as np

N_max = 100000
nsq_counts = np.zeros([N_max, 1])
sq_count = 0

for i in range(N_max):
    number = i
    for div in range(2,int(np.ceil(np.sqrt(number)))):
        if (number % (div*div) == 0):
            sq_count +=1
            break
    nsq_counts[i] = i-sq_count
```

Gęstość liczb bezkwadratowych $\frac{b_n}{n}$ zbiega do $\frac{1}{\zeta(2)}$:

```python
import matplotlib.pyplot as plt
nsq_over_n = [nsq_counts[i]*1./i for i in range(1,N_max)]
plt.plot(range(1,N_max),nsq_over_n,'b')
from scipy.special import zeta
plt.plot([1,N_max],[1/zeta(2),1/zeta(2)],'r')
plt.show();
```

$\frac{b_n}{n}-\frac{1}{\zeta(2)}=f(n)$:

```python
plt.plot(range(1,N_max),np.log10(nsq_over_n-1/zeta(2)),'b')
plt.xlabel('$n$')
plt.ylabel('$\log_{10}(b_n/n-1/\zeta(2))$')
plt.show();
```
